```python
bike = pd.read_csv("bike.csv")
```

```python
bike.head()
```

```python
bike_summer = bike.loc[bike["season"] == 2, ["season", "registered", "datetime"]]
bike_winter  = bike.loc[bike["season"] == 4, ["season", "registered", "datetime"]]
```

```python
bike_summer
```

```python
bike_summer["datetime"] = pd.to_datetime(bike_summer["datetime"]).dt.hour
bike_winter["datetime"] = pd.to_datetime(bike_winter["datetime"]).dt.hour
```

```python
bike_summer
```

```python
bike_summer = bike_summer.groupby(["season", "datetime"]).mean()
bike_winter = bike_winter.groupby(["season", "datetime"]).mean()
```

```python
bike_summer.head()
```

```python
bike_summer
```

```python
bike_summer.reset_index()
bike_winter.reset_index()
```

```python
bike_merge = pd.merge(left=bike_summer, right=bike_winter, left_on="datetime", right_on="datetime", how="inner")
```

```python
bike_merge
```

```python
bike_merge.reset_index()
```

```python
bike_merge["diff"] = abs(bike_merge["registered_x"] - bike_merge["registered_y"])
```

```python
bike_merge
```

```python
bike_merge = bike_merge.reset_index()
```
