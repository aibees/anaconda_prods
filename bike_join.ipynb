```python
import pandas as pd
import numpy as np
```

```python
bike = pd.read_csv("bike.csv")
```

```python
bike.head()
```

```python
bike_h = bike.loc[bike["humidity"] >= 100, ["datetime", "humidity"]]
bike_h.head()
```

```python
bike_h.info()
```

```python
bike_h["datetime"] = bike_h["datetime"].str.slice(0, 10)
```

```python
bike_h = bike_h.groupby("datetime").count()
bike_h
```

```python
bike_h.head()
```

```python
bike_t = bike.loc[bike["temp"] >= 30, ["datetime", "temp", "count"]]
```

```python
bike_t.head()
```

```python
bike_t["ymd"] = bike_t["datetime"].str.slice(0, 10)
```

```python
bike_t
```

```python
bike_merge = pd.merge(left=bike_t, right=bike_h, left_on = "ymd", right_on = "datetime")
```

```python
bike_merge
```

```python
bike_merge["hour"] = pd.to_datetime(bike_merge["datetime"]).dt.hour
```

```python
bike_merge.head()
```
